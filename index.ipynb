```python
import pandas as pd
```

```python
df = pd.read_csv('./data/mc160000.csv', encoding='shift-jis')

df = df.rename(columns={'Unnamed: 0': 'code_cause', 'Unnamed: 1': 'sex'})
df = df.iloc[:, :-4]
df = df.drop(columns=['総数', '0歳', '1歳', '2歳', '3歳', '4歳', '1〜4歳'])
df = df.replace('-', 0)
df['code_cause'] = df['code_cause'].ffill()
df[['code', 'cause']] = df['code_cause'].str.extract(r"(\S+)\s+(.+)")
df['code'] = df['code'].str.strip()
df['cause'] = df['cause'].str.strip()
df = df.loc[df['code_cause'].str.contains('00')]
df = df.loc[~df['code_cause'].str.contains('000')]
df = df.loc[df['sex'] != '総数']
df = df.drop(columns=['code_cause'])
df.set_index('cause', inplace=True)

df
```

```python
def get_age_class(age):
    if age >= 100:
        return '100歳以上'
    else:
        return df.columns[(age // 5) + 1]
```

```python
def get_target(sex, age):
    return df.loc[df['sex'] == sex, get_age_class(age)].astype(float)
```

```python
t = get_target('男', 12)
```

```python
tt = t.sort_values(ascending=False).head(20)
tt
```
